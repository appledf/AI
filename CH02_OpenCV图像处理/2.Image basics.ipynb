```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('image.jpg')
(h,w,c) = image.shape
print(image.shape)
```

```python
(b, g, r) = image[0,0]
print(image[0,0])
```

```python
image[0,0] = (0, 0, 255)
(b, g, r) = image[0,0]
print(image[0,0])
```

```python
cX, cY = (w//2,h//2)
tl = image[0:cY,0:cX]
plt.imshow(tl)
plt.axis('off')
plt.show()
```

```python
tr = image[0:cY, cX:w]
bl = image[cY:h, 0:cX]
br = image[cY:h, cX:w]

def show(image):
    plt.imshow(image)
    plt.axis('off')
    plt.show()
    
show(tl)
show(tr)
show(bl)
show(br)
```

```python
image[0:cY,0:cX] = (0, 0, 255)
show(image)
```
